```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))

sns.boxplot(
    x='target', 
    y='age', 
    data=df, 
    palette=['maroon', '#ff6666'], 
    width=0.6,
    linewidth=2,
    fliersize=5, 
    notch=True 
)

sns.swarmplot(x='target', y='age', data=df, color='black', alpha=0.4, size=4)

plt.title('Comparison of Age Distribution by Heart Disease Status', fontsize=15, fontweight='bold', pad=20)
plt.xlabel('Heart Disease Status (0 = No, 1 = Yes)', fontsize=12, fontweight='semibold')
plt.ylabel('Patient Age', fontsize=12, fontweight='semibold')

plt.xticks([0, 1], ['No Disease', 'Disease'])

plt.show()
```

# 5.piechart


```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('dataset.csv')

target_counts = df['target'].value_counts()
labels = ['Disease', 'No Disease'] 
colors = ['maroon', '#ff6666'] 

plt.figure(figsize=(6, 6))
plt.pie(
    target_counts, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors,
    explode=(0.08, 0), 
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    shadow=True 
)

plt.title('Heart Disease Distribution in Dataset', fontsize=16, fontweight='bold', pad=20)
plt.show()
```

# 6. bar chart


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('dataset.csv')

sns.set_theme(style="white")
plt.figure(figsize=(10, 6))

ax = sns.barplot(
    x='cp', 
    y='thalach', 
    hue='target', 
    data=df, 
    palette=['#ff9999', 'maroon'], 
    edgecolor='black',
    linewidth=1.5,
    capsize=.05
)

plt.title('Max Heart Rate by Chest Pain Type and Disease Status', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Chest Pain Type (0-3)', fontsize=13, fontweight='bold')
plt.ylabel('Average Max Heart Rate (thalach)', fontsize=13, fontweight='bold')

plt.legend(title='Heart Disease', loc='upper left', labels=['No', 'Yes'])
sns.despine()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dataset.csv')

sns.set_theme(style="whitegrid", palette="muted")
plt.figure(figsize=(12, 7))

sns.violinplot(
    x='cp', 
    y='thalach', 
    hue='target', 
    data=df, 
    split=True, 
    inner="quart", 
    palette=['#ff9999', 'maroon'],
    linewidth=1.5
)

plt.title('Heart Rate Distribution by Chest Pain Type and Disease Status', fontsize=18, fontweight='bold', color='maroon', pad=25)
plt.xlabel('Chest Pain Type (0: Typical, 1: Atypical, 2: Non-anginal, 3: Asymptomatic)', fontsize=12, fontweight='bold')
plt.ylabel('Maximum Heart Rate Achieved (thalach)', fontsize=12, fontweight='bold')

plt.legend(title='Heart Disease', loc='upper right', labels=['No', 'Yes'], frameon=True, shadow=True)

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 1. Load the data
df = pd.read_csv('dataset.csv')

# 2. Scale the data to fit inside the panda's face
x_scaled = (df['age'] - df['age'].min()) / (df['age'].max() - df['age'].min()) * 0.6 + 0.2
y_scaled = (df['thalach'] - df['thalach'].min()) / (df['thalach'].max() - df['thalach'].min()) * 0.6 + 0.2

# 3. Create the character layout
fig, ax = plt.subplots(figsize=(8, 8))

# Draw Ears
ax.add_patch(patches.Circle((0.2, 0.8), 0.15, color='black', zorder=1))
ax.add_patch(patches.Circle((0.8, 0.8), 0.15, color='black', zorder=1))

# Draw Face (The main graph area)
ax.add_patch(patches.Circle((0.5, 0.5), 0.45, color='white', ec='black', lw=3, zorder=2))

# Draw Eye Patches
ax.add_patch(patches.Ellipse((0.35, 0.6), 0.2, 0.25, angle=20, color='black', zorder=3))
ax.add_patch(patches.Ellipse((0.65, 0.6), 0.2, 0.25, angle=-20, color='black', zorder=3))

# 4. Plot the Heart Data inside the Panda
plt.scatter(x_scaled, y_scaled, marker='$\heartsuit$', color='maroon', s=40, alpha=0.7, zorder=4)

# 5. Draw the Eyes and Nose
ax.add_patch(patches.Circle((0.35, 0.63), 0.05, color='white', zorder=5))
ax.add_patch(patches.Circle((0.65, 0.63), 0.05, color='white', zorder=5))
ax.add_patch(patches.Polygon([[0.47, 0.4], [0.53, 0.4], [0.5, 0.35]], color='black', zorder=5))

# Formatting
plt.title("The 'Healthy Heart' Panda Analysis ", fontsize=20, fontweight='bold', color='maroon', pad=10)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off') # Hiding axes to show only the character

plt.show()
```
